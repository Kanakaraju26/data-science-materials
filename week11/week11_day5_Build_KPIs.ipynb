# DAX Functions – Simple Notes
## 1️⃣ SUM()

- What it does: Adds up all the numbers in a column.

- Example: SUM(Sales[Sales]) → total sales.

- Simple analogy: Like adding all the money from each bill in your wallet.

## 2️⃣ CALCULATE()

- What it does: Changes the “context” of a calculation.

- Example: CALCULATE([Total Sales], FILTER(...)) → calculate sales but only for certain months, products, or regions.

- Simple analogy: “Calculate only for this specific condition.”

- Think: CALCULATE = Apply a filter before calculating.

## 3️⃣ FILTER()

- What it does: Creates a subset of data that meets a condition.

- Example: FILTER(ALL('Date'[Date]), 'Date'[Date] <= MAX('Date'[Date])) → all dates up to today.

- Simple analogy: “Pick only the rows that match my rule.”

## 4️⃣ ALL()

- What it does: Ignores filters on a column or table.

- Example: ALL('Date'[Date]) → consider all dates, even if a filter is applied.

- Simple analogy: “Look at the full picture, not just what’s selected.”

## 5️⃣ PREVIOUSMONTH()

- What it does: Returns the previous month based on a date column.

- Example: PREVIOUSMONTH('Date'[Date]) → last month’s dates.

- Simple analogy: “Go back one month.”

## 6️⃣ DIVIDE()

- What it does: Divides two numbers safely (avoids errors if denominator = 0).

- Example: DIVIDE([Total Sales] - Previous Month Sales, Previous Month Sales) → MoM % change.

- Simple analogy: “Divide A by B, and don’t break if B = 0.”

## ✅ Extra Note

- DAX is mostly about logic, not memorizing syntax.

- If you understand the purpose of each function, you can look up the exact syntax anytime.

- Over time, using these functions in real projects will naturally make you remember them.
