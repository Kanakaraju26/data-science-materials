# 1️⃣ New DAX Functions
## a) DATEDIFF

- Usage: Calculate difference between two dates.
- Example:
- Delivery Days = DATEDIFF(Sales[Order Date], Sales[Ship Date], DAY)
- Purpose: Get delivery duration in days (used for shipping analysis).

## b) DATEADD

- Usage: Shift the filter context backward/forward by a time period.
- Example:
- Last Month Sales = 
- CALCULATE([Total Sales], DATEADD('DateTable'[Date], -1, MONTH))
- Purpose: Get last month’s value (used for MoM targets & growth).

## c) CALCULATE

- Usage: Change the filter context of a measure.
- Example:
- Last Month Customers =
- CALCULATE([Total Customers], DATEADD('DateTable'[Date], -1, MONTH))
- Purpose: Evaluate measures in different time periods.

## d) DISTINCTCOUNT

- Usage: Count unique values.
- Example:
- Total Customers = DISTINCTCOUNT(Sales[Customer ID])
- Purpose: Get total number of unique customers or orders.

## e) DIVIDE

- Usage: Safe division (avoids divide-by-zero errors).
- Example:
- Average Order Value = DIVIDE([Total Sales], [Total Orders])
- Purpose: Calculate ratios like AOV.

## f) KPI Target Formula (custom example)

- Sales Target = Last Month Sales × 110%
- Sales Target = [Last Month Sales] * 1.10
- Orders Target = Last Month Orders × 105%
- Orders Target = [Last Month Orders] * 1.05
- Customers Target = Last Month Customers × 110%
- Customers Target = [Last Month Customers] * 1.10
- AOV Target = Last Month AOV × 105%
- AOV Target = [Last AOV (MoM)] * 1.05

# 2️⃣ New Charts / Visuals
## a) KPI Card

- Usage: Show key metrics with target & trend.
- Setup:
- Indicator = measure (e.g., [Total Sales])
- Target = target measure (e.g., [Sales Target])
- Trend Axis = 'DateTable'[Month-Year]

## b) Scatter / Bubble Chart

- Usage: Show relationship between two metrics.
- Setup Example:
- X-axis = Delivery Days (calculated column)
- Y-axis = [Total Sales] (proxy for Order Size)
- Bubble size = [Total Orders]
- Legend = Region or Segment

## c) Decomposition Tree

- Usage: Break down a measure step by step.
- Setup Example:
- Analyze = [Total Sales]
- Explain By = Category → Sub-Category → Product Name
- Note: You expand using the “+” button inside the visual.

# ✅ Key Learnings
## MoM vs YoY:

- MoM → short-term operational trends
- YoY → long-term seasonal or strategic trends
- Targets: Built dynamically using Last Period × Growth %.
- Date Dimension Table: Essential for all time-intelligence DAX.
