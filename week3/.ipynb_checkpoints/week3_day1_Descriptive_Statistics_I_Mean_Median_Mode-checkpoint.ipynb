## Descriptive Statistics: Mean, Median, Mode

This dataset represents sample scores of 15 students.

- **Mean** gives the average score.
- **Median** gives the middle score when data is sorted.
- **Mode** gives the most frequent score.

If there's an outlier (like a very high or low value), the **median** is more reliable than the **mean**.
