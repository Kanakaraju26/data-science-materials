# Standard deviation and Variance

## S.D

### Standard deviation is a number that tells you:
### ***How much do the values in my data differ from the average?***

## ***In a Nutshell:***
### Small SD → data is tight around the mean.
### Large SD → data is spread out from the mean.

# Variance 

### Variance is a number that tells you:

### ***"How spread out are the values in my data?"***

## In a Nutshell:
### Variance = average squared distance from the mean
### It’s a measure of spread, but in squared units (e.g., if your data is in dollars, variance is in dollars²).

```python
# Required imports for plotting

import numpy as np
import matplotlib.pyplot as plt
```

```python
# datasets of all frequency (Low medium High)
low_var = [50, 51, 49, 50, 50]
med_var = [40, 50, 60, 45, 55]
high_var = [10, 90, 20, 80, 100]
```

```python
#print(np.std(low_var).round(2))
```

```python
datasets = [low_var,med_var, high_var]
names = ['Low Variance', 'Medium Variance', 'High Variance']
```

```python
for i, data in enumerate(datasets):
    print(f"{names[i]} -> Variance: {np.var(data):.2f}, Std Dev: {np.std(data):.2f}")
```

```python
plt.figure(figsize=(10, 5))

for i, data in enumerate(datasets):
    y = len(data)
    x = np.arange(y)
    plt.plot(x, data, marker='o', label=names[i])

plt.title("Comparing Variance Across Datasets")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
mean = np.mean(data)
std = np.std(data)

plt.plot(data, marker='o')
plt.axhline(mean, color='green', linestyle='--', label='Mean')
plt.fill_between(range(len(data)), mean - std, mean + std, color='green', alpha=0.2, label='±1 Std Dev')
```
