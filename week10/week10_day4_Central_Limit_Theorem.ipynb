# Central Limit Theorem (CLT)

### If you take many random samples from any population (no matter if it’s skewed, uniform, weirdly shaped…), the distribution of the sample means will tend toward a normal distribution (bell curve) as your sample size gets large enough.

## Key points:

- Works even if the population is not normal (e.g., skewed, has outliers).

- Bigger sample sizes make the sample mean distribution closer to a bell curve.

- This is why we can use normal-based statistical tests even on non-normal data.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("data/adult.csv") 
```

```python
# Check skewness visually
sns.histplot(df['capital.gain'], bins=50, kde=True)
plt.show()
```

```python
def sample_means(data, sample_size, num_samples):
    means = []
    for _ in range(num_samples):
        sample = np.random.choice(data, size=sample_size, replace=True)
        means.append(np.mean(sample))
    return means
```

```python
data = df['capital.gain'].values

means_5 = sample_means(data, sample_size=5, num_samples=1000)
means_30 = sample_means(data, sample_size=30, num_samples=1000)
means_100 = sample_means(data, sample_size=100, num_samples=1000)
```

```python
plt.figure(figsize=(12,6))

sns.histplot(means_5, kde=True, color='red', label='n=5')
sns.histplot(means_30, kde=True, color='blue', label='n=30')
sns.histplot(means_100, kde=True, color='green', label='n=100')

plt.legend()
plt.title("Central Limit Theorem Demonstration - Capital Gain")
plt.show()
```

The purpose of the Central Limit Theorem (CLT) isn’t just to show a “cool trick” that things look bell-shaped.
It’s actually the foundation for a lot of modern statistics and data science.

Here’s why it matters so much:

1. It Justifies Using Normal-Based Methods
Many statistical tools (z-scores, t-tests, confidence intervals, ANOVA, regression assumptions) are derived assuming normality.

In the real world, data is rarely perfectly normal — it’s often skewed or has outliers.

CLT says: “If you take averages of large enough samples, the distribution of those averages is normal.”

This lets us apply normal-based formulas to make predictions, test hypotheses, and estimate probabilities even when the original data is ugly.

2. It Powers Hypothesis Testing
When you run an A/B test (e.g., which website design gets more clicks), you’re not testing the raw data — you’re testing the means of your samples.

Thanks to CLT, the difference between sample means follows a predictable normal pattern.

This means we can calculate p-values and make statistically valid decisions.

3. It Enables Confidence Intervals
A confidence interval says:

“We’re 95% sure the true mean lies between X and Y.”
That calculation assumes the sampling distribution is normal.
CLT guarantees this, so we can trust the intervals.

4. It Works in Machine Learning, Too
Many machine learning algorithms (like linear regression, logistic regression, and certain feature importance tests) assume normality of residuals or parameter estimates.

CLT makes those assumptions reasonable when sample sizes are large.

5. In Short — The Big Picture
The CLT bridges messy reality (skewed, noisy, non-normal data)
and clean mathematical tools (normal distribution formulas).
Without CLT, we’d have to build custom probability models for every different data shape — which would be impractical for day-to-day analysis.
