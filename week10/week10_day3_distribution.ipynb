# Normal distribution: 
### Bell curve, mean = median = mode, symmetric.

# Binomial distribution:
### Discrete, two outcomes (success/failure), controlled by n trials and p probability.

# Uniform distribution:
### Equal probability for all values in a range.

# Generating Synthetic Data to Understand Shapes of Distribution

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Normal distribution
data_normal = np.random.normal(loc=50, scale=10, size=1000)
sns.histplot(data_normal, kde=True)
plt.title("Synthetic Normal Distribution")
plt.show()

# Binomial distribution
data_binomial = np.random.binomial(n=10, p=0.5, size=1000)
sns.histplot(data_binomial, kde=False)
plt.title("Synthetic Binomial Distribution")
plt.show()

# Uniform distribution
data_uniform = np.random.uniform(low=0, high=100, size=1000)
sns.histplot(data_uniform, kde=True)
plt.title("Synthetic Uniform Distribution")
plt.show()
```

## Normal Distribution  Bell-shaped curve, most values close to the average, fewer values far from it.
## Binomial Distribution  Counts how many successes occur in a fixed number of yes/no trials.
## Uniform Distribution  All values in a range are equally likely.

```python

df = pd.read_csv("data/adult.csv")

for col in ["age", "hours.per.week"]:
    sns.histplot(df[col], kde=True)
    plt.title(f"{col} Distribution")
    plt.show()
```
