🔹 <b>1. Null Hypothesis (H₀)</b>

- It’s the default assumption — "nothing interesting is happening."

- Example: “Men and women have the same average age in the dataset.”

- You test against this assumption.

🔹 <b>2. Alternative Hypothesis (H₁ or Hₐ)</b>

- The opposite of the null — "something is happening."

- Example: “Men and women have different average ages in the dataset.”

🔹 <b>3. "If the null were true..."</b>

- Means: “Let’s pretend the null hypothesis is correct and see how surprising our data is.”

- Example: If we assume men and women have the same average age, what’s the probability we’d still see the age difference we observed in the sample?

- That probability is the p-value.

🔹 <b>4. False Positive (Type I Error)</b>

- You say there is an effect/difference when in reality there isn’t.

- Example: You conclude "men are older than women" when in truth, there’s no real age difference — you were just fooled by random noise.

🔹 <b>5. False Negative (Type II Error)</b>

- You say there is no effect when there actually is one.

- Example: Men really are older than women in the population, but your test fails to detect it (maybe sample too small).

🔹 <b>6. True Positive</b>

- You correctly detect an effect when it really exists.

- Example: You find a significant difference in average income between education levels, and the difference is real.

🔹 7<b>. True Negative</b>

- You correctly conclude “no effect” when in fact there is no effect.

- Example: You test whether income depends on race in your dataset, and find no difference — and in reality, there is none.

<br><br>
👉 <b>A simple medical test analogy (very common in data science teaching):</b>

- Null hypothesis = patient is healthy.

- Alternative = patient is sick.

- False Positive = test says sick, but actually healthy.

- False Negative = test says healthy, but actually sick.

- True Positive = test says sick, and patient is sick.

- True Negative = test says healthy, and patient is healthy.
