## Relationships define how tables are connected and how filters flow.

- There are two main aspects:

 - Cardinality (type of relationship)
 - Cross-filter direction (single vs both)

### 1. Relationship Cardinality (Types)

a) One-to-Many (1:*) ‚Äî Most Common
Example:
Date table ‚Üí Sales table
One date (e.g., 01-Jan-2017) ‚Üí Many sales transactions for that date.
‚úÖ Works perfectly for fact (transaction) and dimension (lookup) tables.

b) Many-to-One (:1) ‚Äî Same as 1:
Just the reverse direction of above.
Example: Sales table ‚Üí Product table (many sales per one product).
‚ö†Ô∏è Usually Power BI will auto-detect and flip it to 1:*.

c) One-to-One (1:1)
Each row in one table matches exactly one row in the other.
Example:
Employee Table (EmployeeID) ‚Üî EmployeeDetails Table (EmployeeID)
Useful when you split a wide table into two.
‚ö†Ô∏è Rare in reporting models.

d) Many-to-Many (:)
Both sides can have duplicates.
Example:
Students table
Courses table
One student can enroll in many courses, and each course can have many students.
In Power BI, a bridge table (like Enrollment) is usually created.
‚ö†Ô∏è Can lead to ambiguity if not modeled carefully.

### 2. Cross-Filter Direction (Single vs Both)

a) Single Direction (‚Üí)
Filter flows from the lookup (dimension) table to the fact table.
Example:
Product[Category] ‚Üí Sales[ProductKey]
If you filter ‚ÄúCategory = Bikes‚Äù, it limits Sales rows to only Bikes.
‚úÖ Default and safest.

b) Both (‚Üî) [Bidirectional]
Filters flow both ways.
Example:
Sales ‚Üî Product
Filtering Products filters Sales (normal).
Filtering Sales also restricts Products (e.g., show only products that have sales).
‚ö†Ô∏è Risks:
Ambiguous paths if multiple relationships exist.
Can give wrong totals if used unnecessarily.

‚úÖ Best used in:
Many-to-many scenarios (e.g., Students ‚Üî Courses).
Dynamic security (row-level security rules).

### 3. Active vs Inactive Relationships

Active relationship: solid line ‚Üí used by default in DAX.
Inactive relationship: dashed line ‚Üí ignored unless activated with USERELATIONSHIP() in DAX.
Example:
Sales has two dates: OrderDate and ShipDate.
You can relate both to the Date table, but only one can be active.
To use the inactive one:

| Type                 | Example                    | Use Case                     | Notes                 |
| -------------------- | -------------------------- | ---------------------------- | --------------------- |
| 1:\*                 | Date ‚Üí Sales               | Most common (Fact-Dimension) | Default               |
| \*:1                 | Sales ‚Üí Product            | Same as 1:\* but reversed    | Auto-adjusted         |
| 1:1                  | Employee ‚Üî EmployeeDetails | Split tables                 | Rare                  |
| *:*                  | Students ‚Üî Courses         | Many-to-many                 | Needs bridge          |
| Single Direction (‚Üí) | Product ‚Üí Sales            | Safe default                 | Use most of the time  |
| Both Direction (‚Üî)   | Sales ‚Üî Product            | Many-to-many, RLS            | Risky, use sparingly  |
| Active vs Inactive   | OrderDate vs ShipDate      | Multiple date columns        | Use `USERELATIONSHIP` |
