```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("data/airlines_flights.csv")
```

```python
print(df.info())
```

```python
num_cols = ['duration', 'days_left', 'price']
sns.pairplot(df[num_cols])
plt.show()
```

```python
sns.catplot(data=df, x='airline', y='price', kind='box', height=5, aspect=2)
plt.xticks(rotation=45)

sns.catplot(data=df, x='class', y='price', col='source_city', kind='box', height=4)
```

```python
# Price distribution by class
sns.violinplot(data=df, x='class', y='price')

# Price by number of stops
sns.violinplot(data=df, x='stops', y='price')
```

```python
sns.catplot(data=df, x='airline', y='price', hue='class', kind='box', height=6, aspect=2)
```

```python
subset = df[(df['source_city'] == 'Mumbai') & (df['destination_city'] == 'Delhi')]
sns.scatterplot(data=subset, x='days_left', y='price', hue='class')
```

```python
sns.heatmap(df[['duration', 'days_left', 'price']].corr(), annot=True, cmap='coolwarm')
```

#  Seaborn Plot Cheat Sheet: Quick Explanations

## 1. sns.catplot() 
###     Used to compare a numerical value (like price) across categories (like airline, class).
###     Kind of plot can be 'box', 'violin', 'bar', etc.
###   Example: Compare ticket prices across different airlines.

## 2. sns.violinplot() 
###     Shows the distribution (shape + spread) of a numerical value for each category.
###     Similar to boxplot, but includes a density curve.
###   Example: See how flight prices are distributed for Economy vs Business class.

## 3. sns.pairplot() 
###     Creates a grid of scatterplots + histograms for multiple numeric columns.
###     Helps find correlations or patterns between numerical variables.
###     Example: See how price, days_left, and duration relate to each other.

## 4. sns.scatterplot() 
###     Plots individual data points between two numerical variables.
###     Useful for spotting patterns, trends, or outliers.
###     Example: See how price changes with days_left (booking window).

##  scatterplot vs heatmap
### - scatterplot  Shows individual data points (x vs y), raw relationship
### - heatmap  Shows a 2D color grid (often correlation values between numeric features)
### - Example: Use heatmap to check if 'price' is strongly related to 'days_left'


#  Seaborn Plot Syntax Cheat Sheet

## 1. sns.catplot()
###  Use this to compare numeric values across categories (e.g., price by airline).
### Syntax:
### sns.catplot(data=df, x='airline', y='price', kind='box', hue='class')
### - kind: 'box', 'violin', 'bar', 'strip', etc.
### - hue: adds grouping (e.g., split by class)

## 2. sns.violinplot()
###  Shows the distribution (shape + spread) of numeric values per category.
### Syntax:
### sns.violinplot(data=df, x='class', y='price')
### - Combines boxplot + density curve (good for visualizing skewness)

## 3. sns.pairplot()
###  Creates scatterplots and histograms for all pairs of numeric variables.
### Syntax:
### sns.pairplot(df[['price', 'duration', 'days_left']])
### - Optional: add hue='class' to color-code by a category
### - Great for spotting correlation, clusters, outliers

## 4. sns.scatterplot()
###  Plots individual data points between two numeric variables.
### Syntax:
### sns.scatterplot(data=df, x='days_left', y='price', hue='class')
### - Useful for spotting patterns or trends in raw data

## 5. sns.heatmap()
###  Visualizes a matrix (e.g., correlation matrix) using color intensity.
### Syntax:
### sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
### - annot=True: displays numeric values inside each cell
### - cmap: color theme (e.g., 'coolwarm', 'viridis', etc.)
### - Often used to explore feature relationships before modeling
