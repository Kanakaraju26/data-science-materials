```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/HDFC.csv")
```

```python
# print(df.head(10))
print(df.info())
```

```python
df['Date'] = pd.to_datetime(df['Date'])
# df["Date"] = df["Date"].astype('datetime64[ns]')
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df.set_index('Date', inplace=True)
df.sort_index(inplace=True)  # optional but good practice
```

```python
print(df.index.is_unique)
print(df.index.has_duplicates)
print(df.isnull().sum())  # Check for any nulls
```

```python
print(len(df))
print(df.shape)
```

```python
df['Close'].plot(figsize=(12, 6), title='HDFC Daily Closing Prices')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.show()
```

```python
df['Open'].plot(figsize=(12, 6), title='HDFC Daily Open Prices')
plt.xlabel('Date')
plt.ylabel('Open Price')
plt.grid(True)
plt.show()
```

```python
monthly = df['Close'].resample('M').mean()
monthly.plot(figsize=(12, 6), title='Monthly Average Close Price')
plt.xlabel('Month')
plt.ylabel('Average Close Price')
plt.grid(True)
plt.show()
```

```python
print("Original rows:", len(df['Close']))
print("Monthly rows:", len(monthly))
```

```python
weekly_vol = df['Volume'].resample('W').mean()
weekly_vol.plot(figsize=(12, 6), title='Weekly Average Volume')
plt.xlabel('Week')
plt.ylabel('Volume')
plt.grid(True)
plt.show()
```

```python
monthly_oc = df[['Open', 'Close']].resample('M').mean()
monthly_oc.plot(figsize=(12,6), title='Monthly Open vs Close Price')
plt.grid(True)
plt.show()
```
