# 5. Top Countries Producing Netflix Content

```python
top_countries = df['country'].value_counts().nlargest(10)
sns.barplot(x=top_countries.values, y=top_countries.index)
```

# 6. Duration Analysis

```python
sns.histplot(df['duration_mins'], bins=30, kde=True)
sns.boxplot(data=df, x='is_movie', y='duration_mins')
```

#  7. Rating Distribution

```python
ratings_cols = [col for col in df.columns if col.startswith('rating_')]
df['total_ratings'] = df[ratings_cols].sum(axis=1)
rating_sums = df[ratings_cols].sum().sort_values(ascending=False)

sns.barplot(x=rating_sums.values, y=rating_sums.index)
```

#  8. Genre Analysis

```python
from collections import Counter

all_genres = ','.join(df['listed_in'].dropna()).split(',')
genre_counts = Counter([g.strip() for g in all_genres])
genre_df = pd.DataFrame(genre_counts.most_common(10), columns=['Genre', 'Count'])

sns.barplot(data=genre_df, x='Count', y='Genre')
```
